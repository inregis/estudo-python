```python
with open("relatorio.txt", "w") as arquivo:
    arquivo.write("Relatório de Vendas\n")
    arquivo.write("Total: 1500")
 
```

```python
with open("relatorio.txt", "a") as arquivo:
    arquivo.write("\nNovo registro adicionado.")
```

```python
with open("relatorio.txt", "r") as arquivo:
    for linha in arquivo:
        print(linha.strip())
```

```python
with open("dados.txt", "w") as arquivo:
    arquivo.write("Esse é um exemplo de escrita em um arquivo usando Python.\n")
    arquivo.write("Podemos escrever várias linhas de texto.\n")
    arquivo.write("Cada linha é separada por uma quebra de linha.\n")
```
